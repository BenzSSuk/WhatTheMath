# Gradient descent in 2D 
let y = 3x^2 + 5, find x when y = 10 

```python
# set target
target = 10

# define function
def y(x):

    return 3*(x*x) + 5

# Visualize graph
x = np.arange(-1,1,0.1)
plt.plot(x,y(x))
```

## gradient descent iteration

### x_new = x_old + lr*(-dE/dx)
### x_new = x_old - lr*(dE/dx)
### E = target - y

```python
# define E
def E(target,x):
    # E = target - y
    # E = target - ( 3x^2 + 5 )
    # E = -3x^2 -5 + target
    return target - y(x)

def dEdx(x):
    # E = -3x^2 -5 + target
    # dE = 2*(-3)x -0 +0
    return -6*x

# visualize error(Not part in calculation)
x_e = np.arange(-5,5,0.1)
plt.plot(x_e,E(target,x_e)) 
plt.xlabel('x')
plt.ylabel('Error')
```
